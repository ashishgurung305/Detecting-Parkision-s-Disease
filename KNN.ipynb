```python
import numpy as np 
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/srijal/Desktop/Kagle CSV/parkinsons.data')
```

```python
df.corr()
```

```python
features=df.loc[:,df.columns!='status'].values[:,1:]
labels=df.loc[:,'status'].values
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler((-1,1))
x = scaler.fit_transform(features)
y = labels
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2, random_state = 7)
```

```python
from sklearn.metrics import accuracy_score
knn_val = KNeighborsClassifier(n_neighbors = 3)
knn_val.fit(x_train, y_train)
predict_knn = knn_val.predict(x_test)
score_knn = (accuracy_score(y_test, predict_knn)*100)
print('The accuracy is:',score_knn)
```

```python
#creating a confussion matrix

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

confusion_matrix(predict_knn, y_test)

tn,fp,fn,tp = confusion_matrix(predict_knn, y_test).ravel()
(tn,fp,fn,tp)
```

```python
# Predicting the Binary Label:
predict_knn = knn_val.predict(x_test)

# Estimating the probability (likelihood) of Each Label: 
predict_prob_knn = knn_val.predict_proba(x_test)
```

```python
# This line prints the "actual label" of the testing set:
print(y_test)

# This line prints the "predicted label" for the testing set:
print(predict_knn)

# This line prints the "estimated likelihood of both label" for the testing set:
print(predict_prob_knn)

# This line prints the "estimated likelihood of label=1" for the testing set:
print(predict_prob_knn[:,1])
```

```python
from sklearn import metrics

fpr, tpr, thresholds = metrics.roc_curve(y_test, predict_prob_knn[:,1], pos_label=1)

print(fpr)
print(tpr)
```

```python
# AUC:
AUC = metrics.auc(fpr, tpr)
print(AUC)
```

```python
# Importing the "pyplot" package of "matplotlib" library of python to generate 
# graphs and plot curves:
import matplotlib.pyplot as plt

# The following line will tell Jupyter Notebook to keep the figures inside the explorer page 
# rather than openng a new figure window:
%matplotlib inline

plt.figure()

# Roc Curve:
plt.plot(fpr, tpr, color='red', lw=2, 
         label='ROC Curve (area = %0.2f)' % AUC)

# Random Guess line:
plt.plot([0, 1], [0, 1], color='blue', lw=1, linestyle='--')

# Defining The Range of X-Axis and Y-Axis:
plt.xlim([-0.005, 1.005])
plt.ylim([0.0, 1.01])

# Labels, Title, Legend:
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")

plt.show()
```
